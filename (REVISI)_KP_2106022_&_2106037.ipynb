```python
# 10. Analisis Usia dan Diagnosis Hipertensi

plt.figure(figsize=(8, 6))
# Mengganti palet warna menjadi ungu dan kuning
ax = sns.boxplot(x='diag', y='umur', data=df, palette=['purple', 'yellow'])
plt.title('Distribusi Umur Berdasarkan Diagnosis')
plt.xlabel('Diagnosis (1: Hipertensi Primer, 2: Hipertensi Sekunder)')
plt.ylabel('Umur')

# Menampilkan angka pasti pada outlier dan median untuk setiap diagnosis
for diagnosis in df['diag'].unique():
    # Filter data untuk diagnosis tertentu
    subset = df[df['diag'] == diagnosis]

    if not subset.empty:
        # Hitung nilai IQR
        q1 = subset['umur'].quantile(0.25)
        q3 = subset['umur'].quantile(0.75)
        iqr = q3 - q1
        lower_bound = q1 - 1.5 * iqr
        upper_bound = q3 + 1.5 * iqr

        # Ambil outlier
        outliers = subset[(subset['umur'] < lower_bound) | (subset['umur'] > upper_bound)]

        # Tambahkan label untuk setiap outlier
        for index, row in outliers.iterrows():
            # Posisi x untuk diagnosis (disesuaikan untuk diagnosis 1 atau 2)
            x_outlier = diagnosis - 1

            # Tambahkan teks outlier pada plot
            ax.text(x_outlier, row['umur'], f"{row['umur']}", color='black', ha='center', va='bottom')

        # Hitung median
        median_value = int(subset['umur'].median())

        # Posisi x untuk median
        x_median = diagnosis - 1

        # Tambahkan teks median pada plot dengan ukuran lebih kecil dan warna merah
        ax.text(x_median, median_value,
                f"{median_value}",
                color='black',  # Warna putih untuk visibilitas di latar belakang ungu/kuning
                ha='center',
                va='top',
                fontsize=10,  # Ukuran font lebih kecil
                fontweight='bold')  # Berat font normal, tidak tebal

plt.show()
```

### Data Balancing

```python
from imblearn.combine import SMOTEENN
from imblearn.under_sampling import RandomUnderSampler
import pandas as pd

# Misalnya x dan y sudah didefinisikan
# Pertama, lakukan SMOTE-ENN
SE = SMOTEENN(random_state=13)
x_se, y_se = SE.fit_resample(x, y)

# Kemudian, lakukan undersampling
rus = RandomUnderSampler(random_state=13)
x_final, y_final = rus.fit_resample(x_se, y_se)


# Tampilkan jumlah kelas sebelum dan setelah sampling
print('Data Before Sampling:\n', y.value_counts(), "\n")
print('Data After SMOTE-ENN Sampling:\n', pd.Series(y_se).value_counts(), "\n")
print('Data After Undersampling:\n', pd.Series(y_final).value_counts())
```

```python
df['diag'].value_counts().sort_index()
```

## Pemodelan

```python

from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=42)

xse_train, xse_test, yse_train, yse_test = train_test_split(x_se,y_se,test_size=0.2,random_state=42)

xfinal_train, xfinal_test, yfinal_train, yfinal_test = train_test_split(x_final,y_final,test_size=0.2,random_state=42)

# Tampilkan hasil split secara ringkas
print(f"Data x: {x_train.shape[0]} data latih, {x_test.shape[0]} data uji")
print(f"Data x_se: {xse_train.shape[0]} data latih, {xse_test.shape[0]} data uji")
print(f"Data x_final: {xfinal_train.shape[0]} data latih, {xfinal_test.shape[0]} data uji")
```

```python
from sklearn.svm import SVC

# Inisialisasi model dengan C dan gamma
model = SVC(kernel='rbf', C=1.0, gamma=0.1)  # Ganti nilai C dan gamma sesuai kebutuhan

# Latih model menggunakan data pelatihan
model.fit(x_train, y_train.ravel())

# Inisialisasi model dengan C dan gamma untuk data oversampled
model_se = SVC(kernel='rbf', C=1.0, gamma=0.1)  # Ganti nilai C dan gamma sesuai kebutuhan
model_se.fit(xse_train, yse_train.ravel())

# Inisialisasi model dengan C dan gamma untuk data final
model_final = SVC(kernel='rbf', C=1.0, gamma=0.1)  # Ganti nilai C dan gamma sesuai kebutuhan
model_final.fit(xfinal_train, yfinal_train.ravel())
```

```python

#Predict the response for test dataset
y_pred = model.predict(x_test)

yse_pred = model_se.predict(xse_test)
yfinal_pred = model_final.predict(xfinal_test)
```

## Evaluasi

```python
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_pred)
cm_se = confusion_matrix(yse_test, yse_pred)
cm_final = confusion_matrix(yfinal_test, yfinal_pred)

# display(cm)
# display(cm_se)
```
